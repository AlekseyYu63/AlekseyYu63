# Отображение нескольких линейных графиков

Если нужно отобразить на одном графике несколько графиков типа Scatter, можно использовать в качестве параметра метода Plot список из нескольких Scatter:



```
List<double> x1 = new List<double>();
List<double> y1 = new List<double>();
List<double> y2 = new List<double>();
List<double> y3 = new List<double>();
double x0= 0;
for (int i=1; i<100; i++)
{
   
   x1.Add(x0);
   y1.Add(Math.Sin(x0));
   y2.Add(Math.Sin(x0+ Math.PI/3));
   y3.Add(Math.Sin(x0+ Math.PI/2));
   
   x0 +=0.1;
}

for (int i=0; i< x1.Count; i++)
{
 System.Console.WriteLine("  {0:0.000} \t {1:0.000} \t {2:0.000} \t  {3:0.000}",x1[i], y1[i], y2[i], y3[i]);
}
```

Создаем список типа Scatter

```
List<Scatter> список =  new List<Scatter>();
```

Добавляем в него одну линию и отрисовываем график

```
список.Clear();
список.Add(
        new Scatter
         {
    name = "Y1(x)",
    x = x1,
    y = y1,
    mode= "line"
          }
);
Chart.Plot(список).Display();
```

```
список.Clear();
список.Add(
        new Scatter
         {
    name = "Y1(x)",
    x = x1,
    y = y1,
    mode= "line"
          }
);
список.Add(
        new Scatter
         {
    name = "Y2(x)",
    x = x1,
    y = y2,
    mode= "line"
          }
);
список.Add(
        new Scatter
         {
    name = "Y3(x)",
    x = x1,
    y = y3,
    mode= "line"
          }
);
Chart.Plot(список).Display();
```

Либо отрисовываем три линии, используя анонимные (неименованные) объекты

```
var линии =  new List<Scatter> {
     new Scatter
{
    name = "Y1(x)",
    x = x1,
    y = y1,
    mode= "line"
},
     new Scatter
{
    name = "Y2(x)",
    x = x1,
    y = y2,
    mode= "line"
},
     new Scatter
{
    name = "Y3(x)",
    x = x1,
    y = y3,
    mode= "line"
}
    };

Chart.Plot(линии).Display();
```

# Гистограммы
Гистограммы - это не простой столбчатый график. При создании гистограммы происходит автоматическая обработка данных. Например, в следующем примере обрабатывается список из фамилий:
{"Иванов","Петров","Сидоров","Иванов","Иванов","Сидоров","Иванов","Иванов" }
Производится автоматический подсчет количества повторений для каждой фамилии и именно это число - количество повторений фамилии является высотой соостветствующего столбца гистограммы. Столбец "Иванов" - 5, "Петров" - 1, "Сидоров" - 2. 


```



var h = new Histogram
{
    x = new string[] {"Иванов","Петров","Сидоров","Иванов","Иванов","Сидоров","Иванов","Иванов" },
    name = "Оклады"
};

var chart = Chart.Plot(h);
display(chart);
```

```
var h = new Histogram
{
    x = new int[] {0,0,0,0,1,1,1,0,0,3,3,3,3,3,2,0,2,2,10,10,6,7,7,7,7,7,2,2,2,2,2,2 },
    name = "Оклады"
};

var chart = Chart.Plot(h);
display(chart);
```

```
double[] xr = new double[100000];
Random rnd = new Random();
 

for (int i=0; i< 100000; i++)
{
xr[i]=0;
     for (int j=0; j<10; j++)
     {
        xr[i] += 0.1*rnd.Next(); 
     }
         
}


var h = new Histogram
{
    x = xr,
    name = "Оклады"
};

var chart = Chart.Plot(h);
display(chart);
```
